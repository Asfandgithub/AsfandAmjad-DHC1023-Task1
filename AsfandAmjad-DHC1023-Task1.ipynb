### Problem Statement:
#### In the First task , I will explore the famous Iris Dataset And inspect it. In this dataset , measurements of Iris flower are given such

##### Step 1 : Importing Libraries 

```python
import pandas as pd                        #In thsi line I import pandas libraray as pd
import numpy as np                         #In this line I Imported Numpy library as np
import matplotlib.pyplot as plt            #In this line I imported metplot library for plotting graphs 
import seaborn as sns                      #In this line I imported seaborn Library for datavisualisation
```

##### Step 2 : Load the Iris Dataset

```python
iris = sns.load_dataset('Iris')           #Loading Iris dataset from seaborn library using load_dataset
print(iris.head())                        #Using Head Function I tested that It was loaded or not
```

##### Step 3 : Inspecting this Dataset 

```python
iris_head = iris.head()                    #Using head Because no paramete is give it was Printing first 5 rows 
print(iris_head) 
```

```python
iris_tail = iris.tail(8)                    #Using tail printing last 8 rows 
print(iris_tail)
```

```python
#Getting Dataset Size 
iris_size = iris.shape                  #Shape Functions will return rowns and columns size in a dataset (rows,columns)
print(iris_size)                        
```

```python
#Knowing About Labels/column names used in this dataset 
iris_column = iris.columns
print(iris_column)
```

```python
#Knowing About the DataTypes Used in a Data Set and columns are null or not null
iris_info = iris.info()                       #Using Info we will know about Null/nonnullcolumns ,datatypes and Memory
print(iris_info)
```

```python
iris_description = iris.describe()                       #It will show you some stats numbers like mean,minimum,maximum etc.
print(iris_description)
```

##### Step 4 : Data Visualizaton 
##### 1 : Drawing Scatter Plot (Find Relationships Between the variables )

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', s=80)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Sepal Length vs Petal Length by Species')
plt.show()
```

##### 2. Drawing Histogram to see Data Distribution 

```python
plt.hist(iris['sepal_length'],bins=10,color='SkyBlue',edgecolor='Black')
plt.title('Distribution of sepal length ')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

##### 3. Drawing Box Plot for to detect outliers and spread of values 

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
sns.boxplot(data=iris,x='sepal_length')
plt.title('sepal_length')
plt.xlabel('sepal_length')
plt.show()
```
